```python
df1 = df.query("country_name != 'South Sudan' & country_name != 'Congo, Dem. Rep.'")
serie1 = df1.query("series_name == 'Access to electricity (% of population)'").reset_index(drop=True)
years = [col for col in df.columns if col.isnumeric()] 
df2 = pd.melt(serie1, id_vars=["country_name"], value_vars=years, 
                  var_name="year", value_name="percentage")
df2
```

```python
serie_global = df_global.query("series_name == 'Access to electricity (% of population)'").reset_index(drop=True)
years = [col for col in df_global.columns if col.isnumeric()] 
df_global1 = pd.melt(serie_global, id_vars=["country_name"], value_vars=years, 
                  var_name="year", value_name="percentage")
df_global1
```

```python
import geopandas as gpd
world = gpd.read_file("countries/ne_110m_admin_0_countries.shp")

df2['country_name'] = df2['country_name'].replace({'Tanzania': 'United Republic of Tanzania'})
ea_data = df2.merge(world, left_on='country_name', right_on='SOVEREIGNT', how='left')
geo_ea_data = gpd.GeoDataFrame(ea_data, geometry='geometry')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
geo_ea_data.plot(
    column='percentage', 
    cmap='plasma', 
    legend=True, 
    legend_kwds={'label': "Percentage", 'orientation': "horizontal", 'shrink': 0.5, 'pad': 0.02,},
    ax=ax,
    edgecolor='black'
)

ax.set_title('Percentage of Population with Access to Electricity', fontsize=12, fontweight='bold')
ax.axis('off')
plt.tight_layout(pad=2.0)
plt.savefig('map_ea1.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
manual_mapping = {
    "Venezuela, RB": "Venezuela",
    "Congo, Rep.": "Republic of the Congo",
    "United States": "United States of America",
    "Kyrgyz Republic": "Kyrgyzstan",
    "Congo, Dem. Rep.": "Democratic Republic of the Congo",
    "Russian Federation": "Russia",
    "Serbia": "Republic of Serbia",
    "Brunei Darussalam": "Brunei",
    "Korea, Dem. People's Rep.": "North Korea",
    "Iran, Islamic Rep.": "Iran",
    "Eswatini": "eSwatini",
    "Korea, Rep.": "South Korea",
    "Syrian Arab Republic": "Syria",
    "Czech Republic": "Czechia",
    "Slovak Republic": "Slovakia",
    "Cote d'Ivoire": "Ivory Coast",
    "Turkiye": "Turkey",
    "St. Lucia": "Saint Lucia",
    "Bahamas, The": "The Bahamas",
    "Egypt, Arab Rep.": "Egypt",
    "Lao PDR": "Laos",
    "Timor-Leste": "East Timor",
    "Yemen, Rep.": "Yemen",
    "Tanzania": "United Republic of Tanzania",
    "Gambia, The": "Gambia",
    "Micronesia, Fed. Sts.": "Federated States of Micronesia"
}
```

```python
df_global2 = df_global2.copy()
df_global2['matched_country'] = df_global2['country_name'].replace(manual_mapping)
df_global2['final_country'] = df_global2['matched_country'].fillna(df_global2['country_name'])
```

```python
merged_df = world.merge(df_global2, left_on='SOVEREIGNT', right_on='final_country', how='left')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(12, 6))
merged_df.plot(column='percentage', cmap='coolwarm', legend=True, ax=ax)
plt.title("Choropleth Map of Percentages by Country")
plt.tight_layout(pad=2.0)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged_df.plot(
    column='percentage', 
    cmap='Greens', 
    legend=True, 
    legend_kwds={
        'label': "Percentage",
        'orientation': "horizontal",
        'shrink': 0.5,            
        'pad': 0.02,              
    },
    ax=ax,
    edgecolor='black'
)

ax.set_title('Percentage of Population with Access to Electricity', fontsize=12, fontweight='bold')
ax.axis('off')
plt.tight_layout(pad=2.0)
plt.savefig('map5.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
# creating bins and labels
bins = [0, 25, 50, 75, 101]
labels = ["Below 25%", "25% to 49%", "50% to 74%", "Above 75%"]

merged_df["coverage_bin"] = pd.cut(
    merged_df["percentage"],
    bins=bins,
    labels=labels,
    include_lowest=True  # ensures that 0 is included properly
)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged_df.plot(
    column="coverage_bin",
    cmap="viridis", #Greens OrRd viridis plasma coolwarm magma
    categorical=True,
    legend=True,
    legend_kwds={
        "title": "Electricity Coverage",
        "loc": "lower left"
    },
    edgecolor="black",
    ax=ax
)
ax.set_title("Percentage of Population with Access to Electricity", fontsize=12, fontweight="bold", pad=20)
ax.text(
    0.5,
    1.02,
    "Source: The World Bank, 2019",
    ha='center',
    va='bottom',
    transform=ax.transAxes,
    fontsize=10,
)

# annotation text
annotation_text = (
    "While most countries enjoy coverage above 75%, Africa particularly in the sub-Sahara, lags behind with many nations having electricity coverage below 50%."
)

# positioning the annotation
ax.text(
    0.5, -65.0,
    annotation_text,
    ha='center',
    va='top',
    fontsize=10,
    bbox=dict(facecolor="white", alpha=0.8, edgecolor="none")
)

ax.axis("off")
plt.tight_layout(pad=2.0)
plt.savefig('electricity_map1.png', dpi=600, bbox_inches='tight')
plt.show()
```
