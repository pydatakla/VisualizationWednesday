<h3>Kernel Density Estimation</h3>
<p>compute a kernel density estimate (probability density function based on kernels) of a polynormial or function with only one variate</p>
<p>See <a href="https://en.wikipedia.org/wiki/Kernel_density_estimation">Link</a> for details</p>

```python
import seaborn as sn
import seaborn.objects as so
```

```python
seaice = sn.load_dataset('seaice')
```

```python
seaice
```

```python
p = so.Plot(seaice, x="Extent")
p.add(so.Area(), so.KDE())
```

```python
#adjust bandwidth
p.add(so.Area(), so.KDE(bw_adjust=0.23))
```

```python
p2 = p.add(so.Bars(alpha=.3), so.Hist("density"))
p2.add(so.Line(), so.KDE()) 
```

```python
#control the range of the density curve relative to the observations using cut
p2.add(so.Line(), so.KDE(cut=0))
```

```python
#observations assigned to y
so.Plot(seaice, y="Extent").add(so.Area(),
    so.KDE(
    bw_method="silverman"
    ))
```
